#**Publishing and Sharing in Tableau**

1.What is the key difference between publishing a dashboard on Tableau Public
versus Tableau Server/ Cloud in terms of accessibility and data security?

Ans: Tableau Public:

○ Accessible to anyone on the internet
○ No data security (all data is public)
● Tableau Server/Cloud:
○ Access is restricted to authorized users
○ Provides strong data security, authentication, and permissions

2. When publishing a Tableau dashboard, what happens to a live connection
compared to an extract? How does this affect data refresh and performance?

Ans: Live Connection:
● Connects directly to the source database
● Always shows real-time data
● May be slower depending on database performance
Extract:
● Stores a snapshot of data
● Faster performance
● Requires scheduled refresh to update data

3.In Tableau Server or Cloud (conceptually), which feature allows administrators
to control which users can view, interact with, or edit a published dashboard?

Ans: Permissions (User Roles & Groups)
● Admins control who can view, interact, edit, or publish

4.If you want users to view a dashboard but not download the underlying data,
which permission setting would you adjust in Tableau Server?

Ans: Adjust Permissions → Deny "Download Full Data" / "Download Data.

5.Name two methods by which a Tableau dashboard can be embedded into a web
page or website.

Ans: Embed Code (iframe) from Tableau
Tableau JavaScript API

6.List two best practices you should follow when preparing a dashboard for
public sharing to ensure data security and privacy.

Ans: Remove or mask sensitive/confidential data
Use data extracts instead of live connections to avoid exposing databases.

7.After publishing a dashboard, users report a “Data not found” error. What is the
most common cause related to data source settings?

Ans: Most common cause:
● Broken or incorrect data source connection
● (e.g., file path changed, database not accessible.

8.When embedding a Tableau dashboard into a secure internal portal, what two
additional security measures should be implemented to ensure only authorized
users can access the embedded content?

Ans: User Authentication (Login / SSO)
Row-Level Security or Permissions Control.
